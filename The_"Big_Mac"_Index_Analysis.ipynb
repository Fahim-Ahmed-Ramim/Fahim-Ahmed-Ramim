```python
latest_date = df['date'].max()
recent_df = df[df['date'] == latest_date].copy()

# Select specific countries for comparison
countries = ['ARG', 'AUS', 'BRA', 'GBR', 'CAN', 'CHN', 'EUZ', 'HKG', 'IDN', 'JPN', 'MEX', 'SGP', 'ZAF', 'CHE', 'USA']
plot_data = recent_df[recent_df['iso_a3'].isin(countries)].sort_values('USD_raw')
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create the bar chart
colors = ['#ff9999' if x < 0 else '#66b3ff' for x in plot_data['USD_raw']]
sns.barplot(x='USD_raw', y='name', data=plot_data, hue='name', palette=colors, legend=False)

# Formatting
plt.title(f'The Big Mac Index: Currency Valuation vs USD ({latest_date.year})', fontsize=16)
plt.xlabel('Percentage Over/Under Valued (0 = Parity with USD)', fontsize=12)
plt.axvline(0, color='black', lw=1) # Add a line at zero for clarity
plt.tight_layout()
plt.show()
```
